```python
#Importing all pandas packages 
#this way to avoid name space conflicts 
import math as m
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


#one line to open and read a csv filed into a data frame data structure
df3 = pd.read_csv('dem_candidates_clean.csv')

#print the first few lines of the data frame, default is 5 lines
df3.head()
```

After completing lab 4, I now look at my data set and I realize that there are alot of null values which can easily be replaced. 
My next step is to replace these null values with x. 


```python
#filling null values with x through the .fillna method 
df3 = df3.fillna('x')
#print head with "head" method to display replaced null values 
df3.head()
```

Other than replacing those null values, there wasn't much else that needed to be done to clean this data set, as it was pretty well organized. One thing I did do was rename some of the column headers that ended in questions. I simplified these titles to a more consise and snake-case description.

Renaming headers with long names to simplify analysis:

```python
#fixing the headers, some headers have long names

#with the rename method, I called the header name and renamed it

#inplace directly replaces it in your list 
#this makes a permanent change, a view (not just editing the view)

df3.rename(columns = {'LGBTQ?':'lgbtq',
                     'Race' : 'race',
                      'Veteran?' : 'veteran',
                      'Obama Alum' : 'obama'},

              inplace = True)
                      
df3.head()
```

```python
#test to see if it worked (can't see the renamed column headers with the head function)
#so, I called the index of the column through the list function to see if it worked, it did!
print(list(df3)[14])  
```

**Data Modeling:**

I primarily relied on the .count function throughout finding my information, and this served as extremley helpful. In the future, I would love to convert specific parts of my data to a different type in order to do even more analysis. 

**Office Type:**

```python
print("Candidate Information by Office Type They Campaigned For:")

#groupby goes through all of the elements of the listed column as a group 

#the count method automatically counts the amount of every value in the column for the group I have named. 
#So the code below will count the occurance of the different office types in each column!
#I have easy access to super easy, printable information through the groupby and count methods
df3.groupby("office_type").count()

```

Now, I can use this same count method with groupby to find a crazy amount of information for each candidate. Here I am finding the information of how many candidates advanced and how many lost. 

**Wins and Losses Count:**

```python
#win and losses count:
#I can also use the same count method to see how many candidates advanced and how many lost. 
df3.groupby("primary_status").count()
```

What I can do next is even more exciting: I can count how many candidates ran in each state! And what those corresponding states were. The count function counts unique values, so I've completed my goal with the following line of code. 

**Counting Unique States:**

```python
#states 
df3.groupby("state").count()
```

As you can see in my data above, this data set is pretty stable and clean, as the number is constant throughout all of the columns.

**Sorting Candidates by District:**

```python
#I can do the same thing for district and gather more information on 
#how many candidates ran for each district, and I am able to display those unique districts!

df3.groupby("district").count()
```
